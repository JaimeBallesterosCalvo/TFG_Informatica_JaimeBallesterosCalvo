```python
# GUARDAR TODO EN GOOGLE DRIVE - EJECUTAR ESTO AHORA MISMO

import shutil
from google.colab import drive

# Montar Google Drive (si no está montado)
drive.mount('/content/drive')

# Crear carpeta para tu proyecto en Drive
!mkdir -p '/content/drive/MyDrive/TFG_Modelo'

# Copiar el modelo entrenado (LO MÁS IMPORTANTE)
shutil.copy('/content/yolov5/runs/train/my_training/weights/best.pt',
           '/content/drive/MyDrive/TFG_Modelo/best_model.pt')

# Copiar todos los resultados del entrenamiento
shutil.copytree('/content/yolov5/runs/train/my_training',
               '/content/drive/MyDrive/TFG_Modelo/training_results',
               dirs_exist_ok=True)

# Copiar el dataset completo
shutil.copytree('/content/dataset',
               '/content/drive/MyDrive/TFG_Modelo/dataset',
               dirs_exist_ok=True)

# Copiar el código de YOLOv5
shutil.copytree('/content/yolov5',
               '/content/drive/MyDrive/TFG_Modelo/yolov5_code',
               dirs_exist_ok=True)

print("✅ ¡TODO guardado en Google Drive! Revisa la carpeta 'TFG_Modelo' en tu Drive")
```
